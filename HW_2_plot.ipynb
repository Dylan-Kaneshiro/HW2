```python
import pandas as pd, matplotlib.pyplot as plt

dual_values_df = pd.read_csv('dual_values.csv')

# Extract the data for plotting
max_utilization = dual_values_df['Max Utilization']
dual_values = dual_values_df['Dual Value']

# Create a scatter plot
plt.scatter(max_utilization, dual_values)
plt.xlabel('Maximal Utilization of Machines')
plt.ylabel('Dual Values (Shadow Prices)')
plt.title('Maximal Utilization vs Dual Values for Machine Utilization Constraints')
plt.grid(True)
plt.show()
```
